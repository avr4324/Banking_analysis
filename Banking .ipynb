# QUESTION 1 
-	What is the distribution of age among the clients?

```python
plt.figure(figsize = (12,6))
sns.histplot(data['age'], bins = 30, kde = True)
plt.title("Distribution of age among clients")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

# QUESTION 2
-	How does the job type vary among the clients?

```python
plt.figure(figsize = (12,6))
sns.countplot(data['job'], order = data['job'].value_counts().index)
plt.title('Job distribution among Clients')
plt.xlabel("job")
plt.ylabel("Count")
plt.xticks(rotation = '45', ha = 'right')
plt.show()
```
